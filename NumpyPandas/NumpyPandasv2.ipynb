## generar numeros aleatorios pero con default_rg(semilla) REPRODUCIBILIDAD
### RG ES NUESTRO GENERADOR RANDOM DE NUMEROS ENTRE 0 Y 1, ES MUY IMPORTANTE PARA ML, ESTADISTICA ETC

```python
rg = np.random.default_rng(2) # (2)Semillas, para numeros aleatorios
```

### ((2,2)) creacion de una matriz de 2x2

```python
aleatorio = rg.random((2,2)) # ((2,2)) crea una matriz de 2x2
```

```python
aleatorio
```

### ((2,2)) creacion de un arreglo de tamaño 6

```python
azar = rg.random(5) # ((2,2)) crea una arreglo de tamaño 5
```

```python
azar
```

```python
normal = rg.normal(10,5,1000000)
```

```python
normal
```

```python
plt.hist(normal, bins = 100)
plt.show()
```
